```python
# Matplotlib - Axes Class
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes([0.1,0.1,0.8,0.8])
```

```python
# add_axes() function
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
```

```python
# ax.legend() function
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes([0.1,0.1,0.8,0.8])
ax.legend(labels = ('label1','label2'),
          loc = 'upper left')
```

```python
# ax.plot() function
import matplotlib.pyplot as plt
import numpy as np
X = np.linspace(-np.pi,np.pi,15)
C = np.cos(X)
S = np.sin(X)
ax = plt.axes([0.1,0.1,0.8,0.8])
ax1 = ax.plot(X,C, 'bs:')
ax2 = ax.plot(X,S,'ro-')
ax.legend(labels = ('Cosine Function',
                    'Sine Function'),
                    loc = 'upper left')
ax.set_title("Trigonometric Functions")
plt.show()
```
