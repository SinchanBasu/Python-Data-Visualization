```python
# Plot Multiple lines in Matplotlib
# Plotting a single Horizontal Line
import matplotlib.pyplot as plt
x = [10,20,30,40,50]
y = [30,30,30,30,30]
plt.plot(x,y)
plt.show()
```

```python
# Plotting a Single Verical Line
import matplotlib.pyplot as plt
x = [10,20,30,40,50]
y = [30,30,30,30,30]
plt.plot(y,x)
plt.show()
```

```python
# Plotting a Horizontal and Vertical Line
import matplotlib.pyplot as plt
x = [10,20,30,40,50]
y= [30,30,30,30,30]
plt.plot(x,y,label="line 1")
plt.plot(y,x,label="line 2")
plt.legend()
plt.show()
```

```python
# Plotting Multiple Lines
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5]
y = [3,3,3,3,3]

plt.plot(x,y,label = "line 1")
plt.plot(y,x,label = "line 2")
plt.plot(x,np.sin(x),label="curve 1")
plt.plot(x,np.cos(x),label="curve 2")
plt.legend()
plt.show()
```

```python
# Plotting Multiple Lines with different Line Styles
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5]
y = [3,3,3,3,3]

plt.plot(x,y,label="line 1",linestyle="-")
plt.plot(y,x,label="line 2",linestyle="--")
plt.plot(x,np.sin(x),label="curve 1",linestyle="-.")
plt.plot(x,y,np.cos(x),label="curve 2",linestyle=":")
plt.legend()
plt.show()
```
