```python
# How to set the spacing between subplots in Matplotlib in Python?
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
fig,ax = plt.subplots(2,2)

ax[0,0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1,0].plot(x,x*x)
ax[1,1].plot(x,x*x*x)
plt.show()
```

```python
# Using tight_layout() method to set the spacing between subplots
# Without using pad
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
fig,ax = plt.subplots(2,2)

ax[0,0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1,0].plot(x,x*x)
ax[1,1].plot(x,x*x*x)
fig.tight_layout()
plt.show()
```

```python
# Using pad
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
fig,ax = plt.subplots(2,2)

ax[0,0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1,0].plot(x,x*x)
ax[1,1].plot(x,x*x*x)
fig.tight_layout(pad=0.5)
plt.show()
```

```python
# Using subplots_adjust() method to set the spacing between subplots
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
fig,ax = plt.subplots(2,2)

ax[0,0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1,0].plot(x,x*x)
ax[1,1].plot(x,x*x*x)
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```

```python
# Using subplots_tool() method to set the spacing between subplots
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
fig,ax = plt.subplots(2,2)

ax[0,0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1,0].plot(x,x*x)
ax[1,1].plot(x,x*x*x)
plt.subplot_tool()
plt.show()
```

```python
# Using constrained_layout() to set the spacing between subplots
fig,ax = plt.subplots(2,2,
                      constrained_layout = True)
ax[0,0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1,0].plot(x,x*x)
ax[1,1].plot(x,x*x*x)
plt.show()
```
