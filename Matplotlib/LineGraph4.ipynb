```python
# Plot line graph from NumPy array
# Example 1:
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,11)
y = x * x

plt.title("Line graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,color="red")
plt.show()
```

```python
# Example 2:
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,11)
y = np.array([100,10,300,20,500,60,700,80,900,100])

plt.title("Line graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,color="green")
plt.show()
```
